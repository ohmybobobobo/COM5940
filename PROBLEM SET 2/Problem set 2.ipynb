```python
%%time
import requests
import csv
import pandas as pd
from bs4 import BeautifulSoup 
from time import sleep
from random import randint 
from time import time

header = ['url','title','details']
data = []
for i in range(3,7):
    quote_page = requests.get('http://search.kepuchina.cn/search/list?alias=Search&s=健康饮食&page='+ str(i))
    sleep(randint(8,15))
    print("\n***** Page " + str(i) +" in action *****")
    soup = BeautifulSoup(quote_page.content,'html.parser')
                              
    for article in soup.find_all("div", attrs={"class":"item-cont"}): 
        url = article.a.get('href')
        title = article.h5.a.text.encode('utf-8').strip()
        decoded_title = title.decode('utf-8')
        details = article.p.text.encode('utf-8').strip()
        decoded_details = details.decode('utf-8')
        data.append((url,decoded_title,decoded_details))
                              
df = pd.DataFrame(data, 
    columns = header
)
df.to_csv('myjupiterfolder/my_data_1.csv', sep='\t', encoding='utf-8')
```
